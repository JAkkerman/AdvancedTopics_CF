```python
#make a matrix with the weights for a specific time 
def mat_weights(data, row):
    N = 28
    mat = np.zeros((N,N))
    k = 2
    for i in range(N):
        for j in range(N):
            print(data.iloc[row,i])
            mat[i,j] = data.iloc[row,i]
            k += 1
    return mat
```

```python
A = mat_weights(data, 4)
print(A)
```

```python
def vis_network(data,A):
    com_names = ['DBR', 'AMROBK', 'BACR-Bank', 'BNP', 'BYLAN', 'CMZB', 'CSGAG', 'DB',
                 'DZBK', 'ERGBA', 'HSBC', 'HSBC-HSBCBank', 'INTNED', 'LBW', 'NDB',
                 'SANPAO', 'SANTNDR', 'SEB', 'SOCGEN', 'UBS', 'UCBAG', 'BACF-BankNA',
                 'C', 'CRDSUI-USAInc', 'GS', 'JPM', 'MWD', 'RY', 'MIZUHBA', 'NOMURA']
    N = len(com_names)
    plt.figure(figsize=(18,18))
    G = nx.Graph()
    for i in range(N):
        for j in range(i+1,N):
            if (A[i,j] != 0) and (A[i,j] != 1) and ((A[i,j] != 0.99999999 )):
                G.add_edge(com_names[i],com_names[j],weight = A[i,j])
            

    pos = nx.spring_layout(G)  # positions for all nodes

    # nodes
    nx.draw_networkx_nodes(G, pos, node_size=5)

    # edges
    minLineWidth = 1
    for u, v, d in G.edges(data=True):
        d['weight'] = d['weight']*minLineWidth
        
    #edges,weights = zip(*nx.get_edge_attributes(G,'weight').items())
    
    nx.draw_networkx_edges(G,pos,edgelist=G.edges ,alpha=0.5, edge_color='b', style="dashed")
    
    

    # labels
    nx.draw_networkx_labels(G, pos, font_size=15)
    
    plt.axis("off")
    plt.show()
    
    return G
```

```python
G = vis_network(data, A)
```
